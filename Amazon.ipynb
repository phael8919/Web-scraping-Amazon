# WepScraping Produtos no site da Amazon

- Rode a Célula abaixo e digite o nome do produto

```python
produto = input("Digite o nome do produto: ")
#Nome de teste: Notebook 8Gb SSD I5
```

- Rode a célula abaixo para buscar os dados no site da Amazon

```python
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyautogui as tempoEspera
import pandas as pd

navegador = opcoesSelenium.Chrome()
tempoEspera.sleep(4)
navegador.get('https://www.amazon.com.br/')

tempoEspera.sleep(3)
navegador.find_element(By.NAME, 'field-keywords').send_keys(produto)

tempoEspera.sleep(3)
navegador.find_element(By.ID, 'nav-search-submit-button').click()

tempoEspera.sleep(6)
dados = navegador.find_elements(By.CLASS_NAME, 'puis-card-container')
produto = []
preco = []
link = []

for i in dados:
    nome_produto = i.find_element(By.CLASS_NAME, 'a-size-mini').text
    produto.append(nome_produto.upper())

    try:
        preco_produto = i.find_element(By.CLASS_NAME, 'a-price-whole').text
        preco.append(float(preco_produto.replace(".","")))
    except:
        preco_produto = 0
        preco.append(preco_produto)

    link_produto = i.find_element(By.TAG_NAME, 'a').get_attribute('href')
    link.append(link_produto)
    
    #print(f'{nome_produto.upper()} - {preco} - {link}')

#Importando para um Dataframe do Pandas
dados = {
    'Nome do produto': produto,
    'Preço':preco,
    'Link do produto': link
}
tamanho = len(dados['Nome do produto'])
df = pd.DataFrame(dados)
df = df.query('Preço != 0.000')
df
```

- Rode a célula abaixo para gerar uma tabela ordenada por ordem decrescente de preço:

```python
#ordenando por ordem crescente de preço
df.sort_values('Preço')
```

- Caso deseje salvar em Excel, rode a célula abaixo:

```python
df.to_excel('amazon.xlsx', index=False)
```
